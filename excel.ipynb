```python
import pandas as pd
import numpy as np
import openpyxl
import xlwt
from openpyxl import load_workbook
```

## Réception Management Rapport - Mois (30).xlsx


```python
df_asn = pd.read_excel(r'Réception Management Rapport - Mois (33).xlsx', header=None)

asn = df_asn[0]
```

```python
wb = load_workbook('Réception Management Rapport - Mois (33).xlsx') #loading the excel file(Reception Management Rapport - Mois(30).xlsx) 
ws = wb.active # We active the workbook in order for us to make changes or work with the file


ws.delete_rows(idx=1,amount=14) # we delete the first 14 rows of the rapport(idx biscally means the index we want to start to delet the rows and the amount is the amount of rows to be deleted from the index given)

asn = ws['A'] ## select the "A" column from the file
ASNs = []
#creating a loop to extract the values form the A columns
for cell in asn:
    print(f'{cell.value}\n')
    ASNs.append(cell.value)
```

```python
df = pd.DataFrame(ASNs)
```

```python
df.to_excel (r'asn.xlsx', index = None, header=True)
```
